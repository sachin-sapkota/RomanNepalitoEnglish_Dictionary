```python
df = pd.DataFrame(dataset['train'])
```

```python
train_df, val_df = train_test_split(df, test_size=0.1, random_state=42)
```

```python
from datasets import Dataset

train_dataset = Dataset.from_pandas(train_df)
val_dataset = Dataset.from_pandas(val_df)
```

```python
from transformers import T5Tokenizer

tokenizer = T5Tokenizer.from_pretrained('t5-base')
```

```python
task_prefix = "translate English to Python: "
```

```python
def preprocess_function(examples):
    inputs = []
    targets = []
    for ex_intent, ex_rewritten_intent, ex_snippet in zip(
        examples['intent'], examples['rewritten_intent'], examples['snippet']
    ):
        # Skip examples where 'snippet' is None
        if ex_snippet is None:
            continue  # Skip this example
        # Use 'rewritten_intent' if available, else fall back to 'intent'
        if ex_rewritten_intent is not None and ex_rewritten_intent.strip() != '':
            ex_input = ex_rewritten_intent
        elif ex_intent is not None and ex_intent.strip() != '':
            ex_input = ex_intent
        else:
            continue  # Skip if both are None or empty
        inputs.append(task_prefix + ex_input)
        targets.append(ex_snippet)
    # If no valid inputs in this batch, return an empty dict
    if not inputs:
        return {}
    # Tokenize inputs
    model_inputs = tokenizer(
        inputs, max_length=128, truncation=True, padding='max_length'
    )
    # Tokenize targets (labels)
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(
            targets, max_length=128, truncation=True, padding='max_length'
        ).input_ids
    # Replace padding token id's with -100 to ignore them in the loss computation
    labels = [
        [(label if label != tokenizer.pad_token_id else -100) for label in labels_example]
        for labels_example in labels
    ]
    model_inputs['labels'] = labels
    return model_inputs
```

```python
# Map the preprocessing function to the datasets
tokenized_train_dataset = train_dataset.map(
    preprocess_function, batched=True, remove_columns=train_dataset.column_names
)
tokenized_val_dataset = val_dataset.map(
    preprocess_function, batched=True, remove_columns=val_dataset.column_names
)
```

```python
from transformers import T5ForConditionalGeneration

model = T5ForConditionalGeneration.from_pretrained('t5-base')
```

```python
from transformers import Seq2SeqTrainingArguments, Seq2SeqTrainer
```

```python
from transformers import DataCollatorForSeq2Seq

data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)
```

```python
!pip install evaluate
```

```python
from evaluate import load
metric = load('bleu')

def compute_metrics(eval_preds):
    preds, labels = eval_preds
    # Decode the predictions
    decoded_preds = tokenizer.batch_decode(preds, skip_special_tokens=True)
    # Replace -100 in labels with pad_token_id
    labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
    # Decode the labels
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)
    # Strip whitespace
    decoded_preds = [pred.strip() for pred in decoded_preds]
    decoded_labels = [label.strip() for label in decoded_labels]
    # Compute BLEU score
    result = metric.compute(predictions=decoded_preds, references=decoded_labels)
    # Return the BLEU score
    return {"bleu": result["bleu"]}
```

```python
training_args = Seq2SeqTrainingArguments(
    output_dir='./results',
    num_train_epochs=5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    eval_steps=500,
    evaluation_strategy='steps',
    save_steps=1000,
    save_total_limit=2,
    logging_steps=100,
    learning_rate=5e-5,
    load_best_model_at_end=True,
    predict_with_generate=True,
    fp16=torch.cuda.is_available(),  # Use mixed precision if supported
)
```

```python
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train_dataset,
    eval_dataset=tokenized_val_dataset,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)
```

```python
trainer.train()
```

```python
# Load the fine-tuned model and tokenizer for inference
model = T5ForConditionalGeneration.from_pretrained('./code-translator-model')
tokenizer = T5Tokenizer.from_pretrained('./code-translator-model')
```

```python
# Function for inference
def translate_intent_to_code(intent):
    input_text = task_prefix + intent
    input_ids = tokenizer.encode(input_text, return_tensors='pt', truncation=True, max_length=128)
    input_ids = input_ids.to(model.device)
    outputs = model.generate(input_ids, max_length=128, num_beams=5, early_stopping=True)
    code = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return code.strip()

# Example usage
intent = "How do I sort a list of numbers in Python?"
code = translate_intent_to_code(intent)
print("Generated Code:\n", code)
```
